## Import Libraries

```python
import random
import glob
import os
import shutil
import warnings
import numpy
import matplotlib.pyplot as plt
```

## Configurations

```python
IMAGES_FILE_PATH = "./dataset/images_copy/"
LABELS_FILE_PATH = "./dataset/labels/"
CREATE_REQ_FOLDERS = False
```

```python
TRAIN_PATH = "./dataset/train/"
TEST_PATH = "./dataset/test/"
VAL_PATH = "./dataset/validation/"
TRAIN_IMAGES_PATH  = "./dataset/train/images/"
TRAIN_LABELS_PATH = "./dataset/train/labels"
TEST_IMAGES_PATH = "./dataset/test/images/"
TEST_LABELS_PATH = "./dataset/test/labels/"
VAL_IMAGES_PATH = "./dataset/validation/images/"
VAL_LABELS_PATH = "./dataset/validation/labels/"

create_paths_dict = {"train":TRAIN_PATH, "test": TEST_PATH, "validation":VAL_PATH, "train_images": TRAIN_IMAGES_PATH,
                    "train_labels": TRAIN_LABELS_PATH, "test_images":TEST_IMAGES_PATH, "test_labels": TEST_LABELS_PATH,
                    "validation_images": VAL_IMAGES_PATH, "validation_labels": VAL_LABELS_PATH}
```

## Modify Images And Labels

```python
def create_req_folders(paths_dict):
    """
    Creates essential folders
    """
    
    for path in paths_dict.values():
        if os.path.exists(path):
            warnings.warn(f"{path} has been existed before. That folder is not created again. for this reason.")
        else:
            os.mkdir(path)
            print(f"{path} folder was created.")
            

if CREATE_REQ_FOLDERS:
    create_req_folders(create_paths_dict)
```

```python
def remove_not_used_images():
    """
    Removes images that has not two classes.
    """
    images_name = os.listdir(IMAGES_FILE_PATH)
    labels_name = os.listdir(LABELS_FILE_PATH)

    for name in images_name:
        salt_name = name.split(".")[0]
        if salt_name + ".txt" not in labels_name:
            os.remove(IMAGES_FILE_PATH + name)
            print(f"{name} was removed.")
            

remove_not_used_images()
```

```python
def copying(base_folder_path, dest_folder_path, array_names):

    if len(os.listdir(dest_folder_path)) != 0:
        raise ValueError(f"{dest_folder_path} must be empty")

    for name in array_names:
        
        #Copy things
        shutil.copy(os.path.join(base_folder_path, name), os.path.join(dest_folder_path, name))

def split_dataset(train_ratio, val_ratio, test_ratio):
    """
    Splits all images and labels in one folder. 
    """
    if (train_ratio + val_ratio + test_ratio) != 1:
        raise ValueError("Sum of those three parameters must be 1.")
    
    images_name = os.listdir(IMAGES_FILE_PATH)
    labels_name = os.listdir(LABELS_FILE_PATH)
    salts_name = [name.split(".")[0] for name in images_name]
    
    ratios = [train_ratio, val_ratio, test_ratio]
    
    image_pieces = np.split(images_name, np.cumsum(np.multiply(ratios, len(images_name))).astype(int))
    label_pieces = np.split(labels_name, np.cumsum(np.multiply(ratios, len(labels_name))).astype(int))
    
    train_image_names, val_image_names, test_image_names, _ = image_pieces
    train_label_names, val_label_names, test_label_names, _ = label_pieces
    
    # Copy process for train
    copying(IMAGES_FILE_PATH, TRAIN_IMAGES_PATH, train_image_names)
    copying(LABELS_FILE_PATH, TRAIN_LABELS_PATH, train_label_names)
    
    # Copy process for test
    copying(IMAGES_FILE_PATH, TEST_IMAGES_PATH, test_image_names)
    copying(LABELS_FILE_PATH, TEST_LABELS_PATH, test_label_names)
    
    # Copy process for validation
    copying(IMAGES_FILE_PATH, VAL_IMAGES_PATH, val_image_names)
    copying(LABELS_FILE_PATH, VAL_LABELS_PATH, val_label_names)

    
split_dataset(0.5,0.3,0.2)
```
